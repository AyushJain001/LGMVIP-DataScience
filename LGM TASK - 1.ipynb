```python
#data preparation
```

```python
from sklearn import preprocessing
```

```python
X=df.iloc[:,0:4]
X.head()
```

```python
y=df["class"]
y=y.values
y[0:5]
```

```python
#Normalization
```

```python
std=preprocessing.StandardScaler()
X=std.fit_transform(X)
X[0:4]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
#Model development using KNN Algorithm
```

```python
#Finding the best K value for the model
```

```python
best_k=0
best_acc=0.0
acc_arr=np.zeros((9,))
for k in range(1,10):
    knn_temp=KNeighborsClassifier(n_neighbors=k).fit(X_train,y_train)
    y2_result=knn_temp.predict(X_test)
    acc_score=accuracy_score(y_test,y2_result)
    acc_arr[k-1]=acc_score
    if acc_score>best_acc:
        best_k=k
        knn=knn_temp
        best_acc=acc_score
plt.plot(range(1,10),acc_arr)
plt.xlabel("Value of K")
plt.ylabel("Accuracy")
plt.show()
print("The maximum accuracy is ",best_k)
```

# prediction using classification Model

```python
knn=KNeighborsClassifier(n_neighbors=best_k)
knn.fit(X_train,y_train)
y_pred=knn.predict(X_test)
y_pred
```

```python
#Model Evalution
```

```python
acc=accuracy_score(y_pred,y_test)
print("Accuracy of our model is ",str(acc*100),"%")
```

```python
cmat=confusion_matrix(y_test,y_pred)
values=["Iris-setosa","Iris-versicolor","Iris-virginica"]
cmat_val=pd.DataFrame(cmat,columns=values,index=values)
cmat_val
```
